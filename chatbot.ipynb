```python
import os
from groq import Groq

# ── Config ──────────────────────────────────────────────
GROQ_API_KEY = "your_groq_api_key_here"   # ← paste your key
MODEL        = "llama-3.3-70b-versatile"
SYSTEM_PROMPT = "You are a helpful, knowledgeable AI assistant. Answer questions clearly and accurately."

client = Groq(api_key=GROQ_API_KEY)
conversation_history = []   # keeps multi-turn memory

print("✅ Client ready. Model:", MODEL)
```

```python
def chat(user_message: str) -> str:
    """Send a message and get a response (maintains conversation history)."""
    conversation_history.append({"role": "user", "content": user_message})
    
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            *conversation_history
        ],
        temperature=0.7,
        max_tokens=2048,
    )
    
    assistant_msg = response.choices[0].message.content
    conversation_history.append({"role": "assistant", "content": assistant_msg})
    return assistant_msg


def reset():
    """Clear conversation history to start fresh."""
    global conversation_history
    conversation_history = []
    print("🗑️ Conversation reset.")


print("✅ Functions defined.")
```

```python
# ── Single question example ──────────────────────────────
reply = chat("What is the difference between machine learning and deep learning?")
print("🤖", reply)
```

```python
# ── Multi-turn conversation example ─────────────────────
reset()

questions = [
    "Explain quantum computing in simple terms.",
    "Can you give me a real-world use case?",
    "How far are we from practical quantum computers?"
]

for q in questions:
    print(f"\n👤 You: {q}")
    print(f"🤖 Bot: {chat(q)}")
```

```python
# ── Interactive CLI loop (run this cell for live chat) ───
reset()
print("💬 Chatbot started. Type 'exit' to stop, 'reset' to clear history.\n")

while True:
    user_input = input("You: ").strip()
    if not user_input:
        continue
    if user_input.lower() == "exit":
        print("👋 Goodbye!")
        break
    if user_input.lower() == "reset":
        reset()
        continue
    reply = chat(user_input)
    print(f"\nBot: {reply}\n")
```
